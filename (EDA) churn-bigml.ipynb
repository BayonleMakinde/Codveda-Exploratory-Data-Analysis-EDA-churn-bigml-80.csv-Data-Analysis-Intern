```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv(r'C:\Users\Hp\Desktop\churn-bigml-80.csv')
```

Load the Data & Summary Statistics

```python
# Summary Statistics (Mean, Std, Min, Max, Median)
print("--- Summary Statistics ---")
display(df.describe())
```

```python
# Calculating the Mode (most frequent value)
print("\n--- Mode of the dataset ---")
display(df.mode().iloc[0])
```

Data Visualization

```python
# Set up a grid for multiple plots
plt.figure(figsize=(16, 12))
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['Total day minutes'], kde=True, color='skyblue')
plt.title('Distribution of Total Day Minutes')
```

```python
# Boxplot: Customer Service Calls vs Churn
# This shows if people who leave (Churn=True) call customer service more often
plt.figure(figsize=(8, 6))
sns.boxplot(x='Churn', y='Customer service calls', data=df, palette='Set2', hue='Churn', legend=False)

plt.title('Outliers in Customer Service Calls')
plt.show()

plt.show()
```

```python
# Scatter Plot(Day Minutes vs Day Charge)
plt.figure(figsize=(6, 6))
sns.scatterplot(x='Total day minutes', y='Total day charge', hue='Churn', data=df)
plt.title('Relationship: Day Minutes vs Day Charge')
```

```python
# PLOT 3: An elegant Heatmap
import numpy as np

# Define numeric_cols by selecting only columns with numbers
# This is the step that was missing!
numeric_cols = df.select_dtypes(include=[np.number])

# Setup the figure
plt.figure(figsize=(9, 9))

#  Create the mask to hide the redundant upper triangle
# We use numeric_cols.corr() now that it is defined
mask = np.triu(np.ones_like(numeric_cols.corr(), dtype=bool))

# 4Generate the heatmap
sns.heatmap(numeric_cols.corr(), 
            mask=mask, 
            annot=True, 
            fmt=".2f", 
            cmap='turbo', 
            square=True, 
            linewidths=.2, 
            cbar_kws={"shrink": .8})

plt.title('Feature Correlation Matrix', fontsize=15, pad=20)
plt.show()
```
